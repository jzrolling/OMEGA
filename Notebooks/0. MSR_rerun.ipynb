```python
from skimage.transform import rescale, resize

def pole_aware_resize_by_cell(cell, 
                      channel,
                      reorientation=True,
                      nrows=20,
                      ncols=50,
                      npole=4,
                      flatten=True,
                      smoothen=True):
    from skimage import filters
    data = cell.measurements.signal_measurements[channel]['straighten']
    if reorientation:
        half_l = int(data.shape[1]/2)
        if np.average(data[:,:half_l])<np.average(data[:,half_l:]):
            data = np.flip(data,axis=1)
    upscaled = rescale(data,(2,2),anti_aliasing=True)
    l = cell.measurements.morphology_measurements['length']
    pole_length = int(round((0.3/l)*upscaled.shape[1]))
    pole1 = resize(upscaled[:,:pole_length],(nrows,npole),anti_aliasing=True)
    cell_body = resize(upscaled[:,pole_length:-pole_length],(nrows,ncols-2*npole),anti_aliasing=True)
    pole2 = resize(upscaled[:,-pole_length:],(nrows,npole),anti_aliasing=True)
    stitched = np.concatenate([pole1,cell_body,pole2], axis=1)
    if smoothen:
        stitched = filters.gaussian(stitched,sigma=1)
    stitched = stitched/stitched.mean()
    stitched[stitched<0]=0
        
    if flatten:
        return stitched, l
    else:
        return stitched.reshape(nrows,ncols), l


def pole_aware_resize(data,
                      length,
                      channel,
                      reorientation=True,
                      nrows=15,
                      ncols=30,
                      npole=3,
                      flatten=True,
                      smoothen=True):
    from skimage import filters
    if reorientation:
        half_l = int(data.shape[1]/2)
        if np.average(data[:,:half_l])<np.average(data[:,half_l:]):
            data = np.flip(data,axis=1)
    upscaled = rescale(data,(2,2),anti_aliasing=True)
    pole_length = int(round((0.3/length)*upscaled.shape[1]))
    pole1 = resize(upscaled[:,:pole_length],(nrows,npole),anti_aliasing=True)
    cell_body = resize(upscaled[:,pole_length:-pole_length],(nrows,ncols-2*npole),anti_aliasing=True)
    pole2 = resize(upscaled[:,-pole_length:],(nrows,npole),anti_aliasing=True)
    stitched = np.concatenate([pole1,cell_body,pole2], axis=1)
    if smoothen:
        stitched = filters.gaussian(stitched,sigma=1)
    stitched = stitched/stitched.mean()
    stitched[stitched<0]=0
        
    if flatten:
        return stitched.flatten(), length
    else:
        return stitched, length

def regular_resize(cell, 
                   channel,
                   reorientation=False,
                   nrows=15,
                   ncols=30):
    data = cell.measurements.signal_measurements[channel]['straighten']
    if reorientation:
        half_l = int(data.shape[1]/2)
        if np.average(data[:,:half_l])<np.average(data[:,half_l:]):
            data = np.flip(data,axis=1)
    l = cell.measurements.morphology_measurements['length']
    resized = resize(data,(nrows,ncols),anti_aliasing=True)
    resized = om.helper.filters.gaussian(resized,sigma=1)
    resized /= resized.mean()
    return resized.reshape(nrows,ncols), l
```

```python
# convert and standardize img data
folders = sorted(glob.glob(dst_folder+'MSR_PP1*/'))
for f in folders:
    header = f.split('/')[-2]
    if len(glob.glob(f+'*miscellaneous_data.pk'))>0:
        data_dict = pk.load(open(f+'{}_miscellaneous_data.pk'.format(header),'rb'))
        df = pd.read_excel(f+'{}_accepted_cells_morphology.xls'.format(header))
        cell_id = np.array(data_dict['cell_id']).astype(int)
        is_outlier = data_dict['is_outlier']
        lengths = data_dict['lengths']
        straightened = data_dict['FITC']['straightened_axial_data']
        straighten_regularized = [pole_aware_resize(straightened[i],
                                                    lengths[i], channel='FITC')[0] for i in range(len(lengths))]
        straighten_regularized = np.array(straighten_regularized)
        compiled=np.hstack([cell_id,
                            lengths.reshape(-1,1),
                            is_outlier.reshape(-1,1),
                            straighten_regularized])
        np.save('{}{}_GEMATRIA.npy'.format(f, header), compiled)
        print(header)
```

```python
well_stats = []
folders = sorted(glob.glob(dst_folder+'*/'))
for f in folders:
    header = f.split('/')[-2]
    locus = header[-10:]
    excel_file = '{}{}_accepted_cells_FITC.xls'.format(f,header)
    if os.path.isfile(excel_file):
        df = pd.read_excel(excel_file)
        filtered = df[df['is_outlier']==0]
        count = len(filtered)
        mean = np.mean(filtered['mean'].values)
        std = np.std(filtered['mean'].values)
        cv = std/mean
        perc_20 = np.percentile(filtered['mean'].values, 20)
        perc_80 = np.percentile(filtered['mean'].values, 80)
        tail = np.median(np.sort(filtered['mean'])[:20])
        well_stats.append([header,locus,count,mean,std,cv,perc_20,perc_80,tail])
well_stats = pd.DataFrame(well_stats,columns=['key','locus','cell_count','mean_fluorescence',
                                             'std_fluorescence','cv_fluorescence','perc_20','perc_80','tail'])
```

```python
unique, counts = np.unique(well_stats['locus'].values, return_counts=True)
duplicates = unique[np.where(counts>=2)[0]]
counted = []
rep = np.ones(len(well_stats))
for i in range(len(well_stats)):
    locus = well_stats.iloc[i]['locus']
    if locus in duplicates:
        if locus not in counted:
            counted.append(locus)
        else:
            rep[i]=2
well_stats['replicate']=rep
well_stats.to_excel('/Volumes/Extreme_SSD/MSR_rerun_20201112_analysis/well_stats.xls')
```
